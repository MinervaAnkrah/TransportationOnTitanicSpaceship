Some labels are not in the right type. Age should be an integer and VIP should be boolean. These can only be changed after handling the NaN values

```python
train_data.isnull().sum()
```

```python
# NaN values in categorical columns will be replaced with mode and those in numerical columns 
# will be replaced with mean
categorical_columns = train_data.select_dtypes(['object','bool']).columns
```

```python
categorical_columns
```

```python
# cross-checking categorical columns
train_data.head()
```

```python
numerical_columns = train_data.select_dtypes(['float','int']).columns
```

```python
numerical_columns
```

```python
def ReplaceNaN(df):
    cat_col = df.select_dtypes(['object','bool'])
    num_col = df.select_dtypes(['float','int'])
    for i in cat_col:
        df[i] = df[i].fillna(df[i].mode()[0])
    for i in num_col:
        df[i] = df[i].fillna(df[i].mean())
```

```python
ReplaceNaN(train_data)
```

```python
train_data.isnull().sum()
```

```python
train_data['Age'] = train_data['Age'].astype('int64')
train_data['VIP'] = train_data['VIP'].astype('bool')
```

```python
train_data.info()
```

```python
# checking for duplicates in our train dataset
train_data.duplicated().value_counts()
```

```python
train_data.head()
```

```python
# creating a copy of the dataset
train_data_copy = train_data.copy()
```

```python
# removing irrelevant columns 
train_data_copy.drop(['PassengerId','Name'], axis=1, inplace=True)
```

```python
train_data_copy.head()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
le = LabelEncoder()
def EncodingCatColumn(df, le):
    cat_columns = df.select_dtypes(['object','bool']).columns
    le = LabelEncoder()
    for i in cat_columns:
        df[i] = le.fit_transform(df[i])
```

```python
EncodingCatColumn(train_data_copy, le)
```

```python
train_data_copy.head()
```

```python
# spliting dataset into X and y values 
X = train_data_copy.drop('Transported', axis=1)
y = train_data_copy['Transported']
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn import svm
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
cls = svm.SVC(kernel='linear')
```

```python
cls.fit(X_train, y_train)
```

```python
pred = cls.predict(X_test)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
SVM_accuracy = accuracy_score(y_test, pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dcls = DecisionTreeClassifier(max_leaf_nodes= 50)
```

```python
dcls.fit(X_train, y_train)
```

```python
dpred = dcls.predict(X_test)
```

```python
DC_accuracy = accuracy_score(y_test, dpred)
```

```python
print(f"Accuracy of DecisionTreeClassifier is {DC_accuracy}")
print(f"Accuracy of SupportVectorMachine is {SVM_accuracy}")
```
